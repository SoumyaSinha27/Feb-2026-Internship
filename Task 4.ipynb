```python
# Question No: 1
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]

    if amount < 50:
        return "Invalid! Minimum recharge is ₹50"
    elif amount not in valid_plans:
        return "Invalid! Plan not available"
    else:
        return "Recharge Successful!"

while True:
    amt = int(input("Enter recharge amount: "))
    result = validate_recharge(amt)
    print(result)

    if result == "Recharge Successful!":
        break
    else:
        retry = input("Retry? (yes/no): ").lower()
        if retry != "yes":
            break
```

```python
# Question No: 2
def reorder_alert(stock_data):
    for product, qty in stock_data.items():
        if qty < 15:
            print(product, "-> Reorder Alert")
        else:
            print(product, "-> Stock OK")

products = {"Rice": 10,"Sugar": 25,"Oil": 12,"Salt": 18}
reorder_alert(products)
```

```python
# Question No: 3
def student_result(marks):
    total = 0
    for m in marks:
        total += m

    avg = total / len(marks)

    print("Average Marks:", avg)

    if avg >= 50:
        return "Pass"
    else:
        return "Fail"

marks_list = [45, 60, 55, 40]
print("Result:", student_result(marks_list))
```

```python
# Question No: 4
def cab_fare(distance, peak_hour):
    base_fare = 50
    per_km = 12

    total = base_fare + (distance * per_km)

    if peak_hour == "yes":
        total = total + (total * 0.25)

    return total

while True:
    km = float(input("Enter distance in km: "))
    peak = input("Is it peak hour? (yes/no): ").lower()

    fare = cab_fare(km, peak)
    print("Total Fare: ₹", fare)

    retry = input("Want to calculate again? (yes/no): ").lower()
    if retry != "yes":
        break
```

```python
# Question No: 5
def attendance_check(attendance_list):
    present_days = 0

    for day in attendance_list:
        if day == "P":
            present_days += 1

    total_days = len(attendance_list)
    percentage = (present_days / total_days) * 100

    print("Attendance Percentage:", percentage)

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

attendance = ["P", "P", "A", "P", "P", "P", "A", "P"]
print("Status:", attendance_check(attendance))
```

```python
# Question No: 6
def password_checker(password):
    if len(password) < 8:
        return "Weak Password (Minimum length should be 8)"

    has_digit = False
    has_special = False
    special_chars = "@#$"

    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in special_chars:
            has_special = True

    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password (Must contain digit and special character @#$)"

pwd = input("Enter password: ")
print(password_checker(pwd))
```
