```python
employees={"Ravi":75000,"Anita":68000,"Kiran":72000}
emp_name=""
highest_sal=0
for name,salary in employees.items():
    if salary>highest_sal:
        highest_sal=salary
        emp_name=name
print("Highest Salary: ",emp_name,"-",highest_sal)
```

```python
prices=[450,1200,899,1500,300]
count=0
for price in prices:
    if price>1000:
        count=count+1
print("Products above 1000: ",count)
```

```python
attendance=["P","P","A","P","P"]
count=0
for attend in attendance:
    if attend=="P":
        count+=1
percentage=(count/len(attendance)*100)
print("Attendance percentage= ",percentage)
```

```python
text=input("Enter String: ")
freq={}
for ch in text:
    if ch in freq:
        freq[ch]+=1
    else:
        freq[ch]=1
print(freq)
```
